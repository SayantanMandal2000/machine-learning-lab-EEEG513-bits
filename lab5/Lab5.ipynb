Import the sklearn library.

```python
import sklearn
```

Load the iris dataset using the “load” command from the sklearn library.

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

Save the predictor variables from the dataset to a variable X

```python
X = iris.data
```

Display the name of the predictor variables.

```python
print("Predictor variable names:", iris.feature_names)
```

Save the target variables from the dataset to a variable y

```python
y = iris.target
```

Use the decision tree classifier from the sklearn library and provide split criterion as “Gini-impurity”.

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
# Create Decision Tree Classifier with Gini impurity
clf = DecisionTreeClassifier(criterion="gini", random_state=42)
```

Display the flow-chart/plot of the final built decision tree on the entire iris dataset.

```python
import matplotlib.pyplot as plt
# Fit the model
clf.fit(X, y)
# Plot the decision tree
plt.figure(figsize=(18, 10))
plot_tree(clf,
          feature_names=iris.feature_names,
          class_names=iris.target_names,
          filled=True,
          rounded=True,
          fontsize=10)
plt.title("Decision Tree for Iris Dataset (Gini Impurity)")
plt.show()
```

Test the model with sample value [5, 3.5, 1.3, 0.3] using the “predict” command of the sklearn library and display the output.

```python
# Test sample
sample = [[5, 3.5, 1.3, 0.3]]
prediction = clf.predict(sample)

# Display result
print("Predicted class (numeric):", prediction[0])
print("Predicted class (species):", iris.target_names[prediction[0]])
```

```python
from sklearn.datasets import load_wine
from sklearn.tree import DecisionTreeClassifier, plot_tree
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wine = load_wine()
```

```python
X = pd.DataFrame(data=wine.data, columns=wine.feature_names)

print("Predictor variable names:")
print(X.columns.tolist())
```

```python
y = wine.target
```

f) Use the decision tree classifier from the sklearn library and provide the split criterion as “Entropy”.

g) Display the flow-chart/plot of the final built decision tree on the entire wine dataset.
(Note: Use the entire wine data to build the decision tree and do not use any data splitting commands.)


```python
clf = DecisionTreeClassifier(criterion="entropy", random_state=42)
clf.fit(X, y)
```

```python
plt.figure(figsize=(20, 10))
plot_tree(clf,
          feature_names=wine.feature_names,
          class_names=wine.target_names,
          filled=True,
          rounded=True,
          fontsize=9)
plt.title("Decision Tree on Entire Wine Dataset (Entropy)")
plt.show()
```
